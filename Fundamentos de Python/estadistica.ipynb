## 1.1 Distribuciones de frecuencia

### 1.1.1 Frecuencias en variables cuantitativas

### 1.1.2 Frecuencias en variables cualitativas


```python
pd.DataFrame({"Frecuencia":data["Sexo"].value_counts(),
              "Porcentaje":data["Sexo"].value_counts(normalize=True) # Al normalizar se da el porcentaje de los valores
            })
```

```python
# Tabla de cruce
pd.crosstab(data["Sexo"], data["Color"]).rename(index={
                                                                0:"Masculino", 1:"Femenino"},
                                                                columns={   0: "Indigena",
                                                                            2: "Blanca",
                                                                            4: "Negra",
                                                                            6: "Amarilla",
                                                                            8: "Marrén",
                                                                            9: "NA"})
```

```python
# Tabla de cruce
pd.crosstab(data["Sexo"], data["Color"], aggfunc="mean", values=data["Renta"]).rename(index={
                                                                0:"Masculino", 1:"Femenino"},
                                                                columns={   0: "Indigena",
                                                                            2: "Blanca",
                                                                            4: "Negra",
                                                                            6: "Amarilla",
                                                                            8: "Marrén",
                                                                            9: "NA"})
```
