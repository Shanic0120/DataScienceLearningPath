# Codificación nominal

Una variable categórica nominal, a diferencia de las ordinales, 

```python
import pandas as pd
```

```python
data = pd.DataFrame({
    "nombre": ["Juan", "Mariana", "Luisa", "Maria", "Paula", "Andres", "Luis", "Camila", "José"],
    "saldo": [10000, 8000, 3000, 4500, 8000, 9000, 5600, 5500, 1000],
    "pais": ["Argentina", "Bolivia", "Chile", "Ecuador", "Colombia", "Costa Rica", "Cuba", "Perú", "Perú"]
})
data
```

```python
data["pais"] = data["pais"].astype("category")
data.dtypes
```

## OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
enc = OneHotEncoder()
df_onehot = pd.concat([data.drop("pais", axis=1), pd.DataFrame(enc.fit_transform(data[["pais"]]).toarray(), columns=enc.categories_)], axis=1)
df_onehot
```
