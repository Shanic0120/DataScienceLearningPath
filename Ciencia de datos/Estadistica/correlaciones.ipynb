# Correlación

Una correlación explica la relación entre dos variables, nos indica la magnitud y dirección de esta relación.

> Correlación NO implica causalidad

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
people = pd.read_csv("../../data/raw/personas.csv")
people.head()
```

```python
sns.pairplot(people)
plt.show()
```

## Pearson

$$r_{xy}=\frac{\sum Z_x Z_y}{N}$$

Donde:

$x, y:$ Variables a relacionar

$Z:$ Desviación estandar

$N:$ Numero de datos

```python
people.corr(method="pearson")
```

```python
sns.heatmap(people.corr(method="pearson"), annot=True, cmap="coolwarm")
plt.tight_layout()
```
